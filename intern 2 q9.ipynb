```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://internshala.com/fresher-jobs')
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify)
```

```python
title=soup.find_all('div',class_='heading_4_5 profile')
title
```

```python
Job_title=[]
for i in title:
    Job_title.append(i.text.strip())
Job_title
```

```python
company=soup.find_all('div',class_='heading_6 company_name')
company
```

```python
Company_name=[]
for i in company:
    Company_name.append(i.text.strip())
Company_name
```
