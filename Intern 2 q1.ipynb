```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://en.wikipedia.org/wiki/Main_Page%E2%80%99')
page
```

```python
page.content
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify)
```

```python
tags=soup.find_all('h1',class_='firstHeading')
tags
```

```python
H1_tags=[]
for i in tags:
    H1_tags.append(i.text.strip())
H1_tags
```

```python
tags=soup.find_all('div',class_='vector-menu-content')
tags
```

```python
h3_tags=[]
for i in tags:
    h3_tags.append(i.text.strip())
h3_tags
```

```python
H3_tags=[]
for i in tags:
    H3_tags.append(i.text.strip('\n'))
H3_tags
```

```python
import pandas as pd
Tags=pd.DataFrame({})
Tags['tags']=H1_tags
Tags
```

```python
import pandas as pd
H3Tags=pd.DataFrame({})
H3Tags['tags']=H3_tags
H3Tags
```
