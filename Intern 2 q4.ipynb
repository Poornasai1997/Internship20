```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://bookpage.com/reviews')
```

```python
page.content
```

```python
script=BeautifulSoup(page.content)
script
```

```python
print(script.prettify)
```

```python
Name=script.find_all('h4',class_='italic',)
Name
```

```python
Book_name=[]
for i in Name:
    Book_name.append(i.text.strip())
Book_name
```

```python
Name=script.find_all('p',class_='sans bold')
Name
```

```python
Author_name=[]
for i in Name:
    Author_name.append(i.text.strip())
Author_name
```

```python
Genre=script.find_all('p',class_='sans bold')
Genre
```

```python
Book_genre=[]
for i in Genre:
    Book_genre.append(i.text.strip())
Book_genre
```

```python
Review=script.find_all('p',class_='excerpt')
Review
```

```python
Book_review=[]
for i in Review:
    Book_review.append(i.text.strip())
Book_review
```

```python
import pandas as py
Bookinformation=py.DataFrame({})
Bookinformation['Name']=Book_name
Bookinformation['Name']=Book_name
Bookinformation['Genre']=Book_genre
Bookinformation['Review']=Book_review
```
